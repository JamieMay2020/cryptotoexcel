```python
def hourly_data(coins):
    columnz = ['Symbol', 'Price','Price Two-Hour Ago', 
               'Two-Hour Percentage Change','Price One-Day Ago', 
               'One-Day Percentage Change','Price One-Month Ago', 
               'One-Month Percentage Change', 'Two-Hour Return Percentile',
               'One-Day Return Percentile','One-Month Return Percentile', 'HQM Score']
    
    final_df = pd.DataFrame(columns = columnz)
    for symbol in coins:
        new_frame = pd.DataFrame(columns=columnz)
        frame = pd.DataFrame(client.get_historical_klines(symbol, '2h', '744 hours ago UTC'))
        frame = frame[[0,4]]
        frame.columns = ['Time', 'Close']
        frame.Close = frame.Close.astype(float) 
        frame.Time = pd.to_datetime(frame.Time, unit='ms')
        new_frame = working(frame, symbol)
        final_df = final_df.append(new_frame)
        
    final_df.reset_index(inplace=True)
    final_frame = percentile(final_df)
    final_df = HQM(final_frame)
    
    final_frame.sort_values(by = 'HQM Score', ascending=False, inplace=True)
    final_frame.drop('index',axis=1, inplace=True)


    final_frame.reset_index(inplace=True)
    column_to_move = final_frame.pop("HQM Score")
    final_frame.insert(2, "HQM Score", column_to_move)

    return final_frame
        
```

```python
def working(frame, symbol):
    columnz = ['Symbol', 'Price','Price Two-Hour Ago', 
               'Two-Hour Percentage Change','Price One-Day Ago', 
               'One-Day Percentage Change','Price One-Month Ago', 
               'One-Month Percentage Change', 'Two-Hour Return Percentile',
               'One-Day Return Percentile','One-Month Return Percentile', 'HQM Score']
    
    new_frame = pd.DataFrame(columns=columnz)
    two_change = round((frame.iloc[-1]['Close']-frame.iloc[-2]['Close'])/frame.iloc[-2]['Close']*100, 4)
    day_change = round((frame.iloc[-1]['Close']-frame.iloc[-12]['Close'])/frame.iloc[-12]['Close']*100, 4)
    month_change = round((frame.iloc[-1]['Close']-frame.iloc[0]['Close'])/frame.iloc[0]['Close']*100,4)
    price_close = frame[:-1]['Close']
    info = client.get_symbol_info(symbol)
    name = info['baseAsset']
    av = (two_change+day_change+month_change)/3
    new_frame = new_frame.append(
    pd.Series([
            name,
            frame.iloc[-1]['Close'],
            frame.iloc[-2]['Close'],
            two_change,
            frame.iloc[-12]['Close'],
            day_change,
            frame.iloc[0]['Close'],
            month_change,
            0,
            0,
            0,
            'N/A'],
        index = columnz),ignore_index= True)
    return new_frame
```

```python
def percentile(final_frame):
    from scipy.stats import stats
    time_periods = ['Two-Hour','One-Day','One-Month']
    for row in final_frame.index:
        for time in time_periods:
            final_frame.loc[row, f'{time} Return Percentile'] = stats.percentileofscore(final_frame[f'{time} Percentage Change'], final_frame.loc[row, f'{time} Percentage Change'])/100
    return final_frame
            
            
            
```

```python
def HQM(df):
    time_periods = ['Two-Hour','One-Day','One-Month']
    for row in df.index:
        momentum=[]
        for time in time_periods:
            momentum.append(df.loc[row, f'{time} Return Percentile'])
            
        df.loc[row, 'HQM Score'] = mean(momentum)
    return df
```
